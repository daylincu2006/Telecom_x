```python
# 2.Crear un nuevo DataFrame con columnas especÃ­ficas
df_tipo_contrato = df_clientes[['Churn', 'account.Contract']]
df_tipo_contrato

# Mostrar el nuevo DataFrame
print(df_tipo_contrato)
```

```python
counts = df_tipo_contrato.groupby(['account.Contract', 'Churn']).size().unstack()
ax = counts.plot(kind='bar', stacked=True, colormap='Set2', figsize=(10,6))
ax.set_title('DistribuciÃ³n del Churn segÃºn Tipo de Contrato')
ax.set_xlabel('Tipo de Contrato')
ax.set_ylabel('Cantidad de Clientes')
ax.legend(title='Churn')

plt.xticks(rotation=0)
plt.show()
```

```python
# 3.Crear un nuevo DataFrame con columnas especÃ­ficas
df_metodo_pago = df_clientes[['Churn', 'account.PaymentMethod']]
df_metodo_pago

# Mostrar el nuevo DataFrame
print(df_metodo_pago)
```

```python
# Agrupar y contar
proporciones = df_metodo_pago.groupby(['account.PaymentMethod', 'Churn']).size().reset_index(name='counts')

# Calcular totales por mÃ©todo de pago
totales = df_metodo_pago.groupby('account.PaymentMethod').size().reset_index(name='total')

# Unir para calcular porcentajes
proporciones = proporciones.merge(totales, on='account.PaymentMethod')
proporciones['percentage'] = (proporciones['counts'] / proporciones['total']) * 100

# Visualizar con seaborn
sns.barplot(
    data=proporciones,
    x='account.PaymentMethod',
    y='percentage',
    hue='Churn'
)

plt.title('ProporciÃ³n de Churn por MÃ©todo de Pago')
plt.xlabel('MÃ©todo de Pago')
plt.ylabel('Porcentaje (%)')
plt.xticks(rotation=45)

plt.show()
```

```python
# 4.Crear un nuevo DataFrame con columnas especÃ­ficas
df_permanencia = df_clientes[['Churn', 'customer.tenure','account.Charges.Monthly']]
df_permanencia

# Mostrar el nuevo DataFrame
print(df_permanencia)
```

```python

# Crear rangos de permanencia
bins = [0, 12, 24, 36, 48, 60, np.inf]
labels = ['0-12', '13-24', '25-36', '37-48', '49-60', 'mÃ¡s de 60']
df_permanencia.loc[:, 'tenure_range'] = pd.cut(df_permanencia['customer.tenure'], bins=bins, labels=labels)

# Contar por rango y churn
counts = df_permanencia.groupby(['tenure_range', 'Churn']).size().unstack()

# Plotear con estilo atractivo
ax = counts.plot(kind='bar', stacked=True, colormap='viridis', figsize=(10,6))
ax.set_title('Clientes segÃºn rango de permanencia y churn')
ax.set_xlabel('Rango de permanencia (meses)')
ax.set_ylabel('Cantidad de clientes')
plt.xticks(rotation=45)
plt.legend(title='Churn')
plt.show()
```
